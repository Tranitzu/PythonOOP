```python
var1 = 'globalny'

def sample_function():

    var1 = 'lokalny funkcji sample_function/obejmujący inner_function'
    print(var1)

    def inner_function():

        var1 = 'lokalny funkcji inner_function'
        print(var1)

    inner_function()

    print(var1)

print(var1)
sample_function()
print(var1)
```

###Instrukcja nonlocal

```python
def sample():

    num1 = 10

    def inner():
        print(num1)

    inner()

sample()
```

```python
def sample():

    num1 = 10

    def inner():
        num1 = 20
        print(num1)

    print(num1)
    inner()
    print(num1)

sample()
```

```python
def sample():

    num1 = 10

    def inner():
        nonlocal num1
        num1 = 20
        print(num1)

    print(num1)
    inner()
    print(num1)

sample()
```
